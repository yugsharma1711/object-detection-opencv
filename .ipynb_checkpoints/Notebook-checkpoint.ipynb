```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_path = './Model/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = './Model/frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_path)
```

```python
classlabels = {}
file_name = './Labels/Lables.txt'
with open(file_name,'rt') as fpt:
    classlabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classlabels)
```

```python
print(len(classlabels))
```

```python
image = cv2.imread('./Test_data/car_4.jpeg')
plt.imshow(image)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5, 127.5,127.5))
model.setInputSwapRB(True)
```

```python
Classindex, confidence, bbox = model.detect(image, confThreshold = 0.6)
```

```python
model.detect(image)
Classindex
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for c_index, conf, boxes in zip(Classindex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(image, boxes, (255,0,0), 2)
    cv2.putText(image, classlabels[c_index-1], (boxes[0] + 10,boxes[1]+40), font, fontScale = font_scale,color = (0, 255,0), thickness = 1)
```

```python
plt.imshow(image)
```

```python
cap = cv2.VideoCapture('./Test_data/Pexels Videos 4698.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Video not opened')

    
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
while True:
    ret, frame = cap.read()
    Classindex, confidence, bbox = model.detect(frame, confThreshold = 0.55)
    print(Classindex)
    if len(Classindex) != 0:
        for c_index, conf, boxes in zip(Classindex.flatten(), confidence.flatten(), bbox):
            if c_index <= 80 :
                cv2.rectangle(frame, boxes, (255,0,0), 2)
                cv2.putText(frame, classlabels[c_index-1], (boxes[0] + 10,boxes[1]+40), font, fontScale = font_scale,color = (0, 255,0), thickness = 1)
    cv2.imshow('Object Detection', frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
